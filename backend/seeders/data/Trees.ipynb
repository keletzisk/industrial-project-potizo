```python
import pandas as pd
```

# READ ALL CSV files into a dataframe


```python
# READ ALL CSV files into a dataframe

# List the filenames of the CSV files
filenames = ['ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - 40 ΕΚΚΛΗΣΙΕΣ.csv',
             "ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - Α' ΔΗΜ. ΔΙΑΜ. - ΙΣΤ. ΚΕΝΤΡΟ.csv",
             "ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - Β' ΔΗΜ. ΔΙΑΜ..csv",
             "ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - Γ' ΔΗΜ. ΔΙΑΜ..csv",
             "ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - Δ' ΔΗΜ. ΔΙΑΜ..csv",
             "ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - Ε' ΔΗΜ. ΔΙΑΜ..csv",
             "ΘΕΣΕΙΣ ΦΥΤΕΥΣΕΩΝ ΤΕΛΙΚΟ ΓΙΑ ΑΝΕΒΑΣΜΑ.xlsx - ΤΡΙΑΝΔΡΙΑ.csv"]

# Create an empty list to store the DataFrames
dfs = []

# Iterate over the filenames and read each CSV file into a DataFrame
for filename in filenames:
    df = pd.read_csv(filename)
    dfs.append(df)

# Concatenate the DataFrames into a single DataFrame
df = pd.concat(dfs, ignore_index=True)

# Count the number of rows
row_count = df.shape[0]

# Print the count
print("Number of rows:", row_count)
```

```python
df.head()
```

# Clean

```python
# Define the value to be deleted
value_to_delete = 'Υπουργου Φιλ. Δραγούμη 1-6'

# Create a boolean mask for rows where 'geolocation' has the specified value
mask = df['GEOLOCATION'] == value_to_delete

# Delete the rows where 'geolocation' has the specified value
df = df[~mask]

# Reset the index of the DataFrame after deleting rows
df = df.reset_index(drop=True)
```

```python
# Filter the rows where the 'GEOLOCATION' column is empty
filtered_df = df[df['GEOLOCATION'].isnull()]

# Display the filtered DataFrame
print(filtered_df)
```

```python
# Delete the rows where the 'GEOLOCATION' column is empty
df = df.dropna(subset=['GEOLOCATION'])

# Reset the index of the DataFrame after deleting rows
df = df.reset_index(drop=True)
```

```python
# Filter the rows where the 'ΕΙΔΟΣ' column contains a forward slash
filtered_df = df[df['ΕΙΔΟΣ'].str.contains('/')]

# Display the filtered DataFrame
print(filtered_df)
```

```python
# Remove the rows where the 'ΕΙΔΟΣ' column contains a forward slash
df = df[~df['ΕΙΔΟΣ'].str.contains('/')]

df = df[~df['ΕΙΔΟΣ'].str.contains(',')]
```

```python

# Trim leading and trailing spaces from all columns
df = df.applymap(lambda x: x.strip() if isinstance(x, str) else x)
```

```python
# Count the number of rows
row_count = df.shape[0]

# Print the count
print("Number of rows:", row_count)
```

# Merge same Types

```python
# Replace the wrong species
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('1 Celtis', 'Celtis australis')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Acer pesudoplatanus', 'Acer pseudoplatanus')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Liquidamparis orientalis', 'Liquidambaris orientalis')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Καλ. Δαμασκηνιά', 'Καλλωπιστική Δαμασκηνιά')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Καλλωπιστική δαμασκηνιά', 'Καλλωπιστική Δαμασκηνιά')

df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Acer campestre', 'Σφένδαμος πεδινός')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Acer platanoides', 'Σφένδαμος ο πλατανοειδής')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Acer pseudoplatanus', 'Σφενδάμι ψευδοπλάτανος')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Celtis australis', 'Κέλτις')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Celtis', 'Κέλτις')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Fotinia', 'Δενδρώδης φωτίνια')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Δενδρώδης Fotinia', 'Δενδρώδης φωτίνια')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Fraxinus ornus', 'Φράξος')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Ginko biloba', 'Γκίνγκο')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Liquidambaris orientalis', 'Υγράμβαρη')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Magnolia grandiflora', 'Μανόλια δενδρώδης')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Robinia pseudacacia', 'Ψευδοακακία')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Δρυς', 'Δρυς πλατύφυλλη')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Κυπαρίσσι', 'Κυπαρίσσι ορθόκλαδο')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Μανόλια', 'Μανόλια δενδρώδης')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Σφαιρική Ακακία', 'Ψευδοακακία')
df['ΕΙΔΟΣ'] = df['ΕΙΔΟΣ'].replace('Sorbus aucuparia', 'Αγριοσορβιά')

```

# Split GEOLOCATION

```python
# Split the 'geolocation' column into 'latitude' and 'longitude'
df[['latitude', 'longitude']] = df['GEOLOCATION'].str.split('[, ]+', 1, expand=True)
# Delete the 'geolocation' column
df = df.drop('GEOLOCATION', axis=1)
```

```python
# List all unique values in the 'ΕΙΔΟΣ' column
unique_values = df['ΕΙΔΟΣ'].unique()

# Display the unique values
for value in unique_values:
    print(value)
```

```python
# Count the rows for each unique type of tree
tree_counts = df['ΕΙΔΟΣ'].value_counts()

# Print the count of rows for each unique type of tree
print(tree_counts)
```

# Create the id and the other columns

```python
# Calculate the length of the DataFrame
length = len(df)

# Create the 'id' column with values starting from 50001
df['id'] = range(50001, 50001 + length)
```

```python
# Create the 'type' column with the value 'Tree' and the id of each row
df['name'] = 'Tree' + df['id'].astype(str)
```

```python
# Create the 'needs_watering' and 'is_alive' columns with values set to "0"
df['needs_watering'] = "0"
df['is_alive'] = "0"
```

```python
df['zip'] = "ΑΓΝΩΣΤΟΣ"
```

```python
df['creation_date'] = "2023-01-01"
```

```python
df['last_watered_date'] = "00:00:00"
```

```python
df.head()
```

```python
# Rename the 'ΕΙΔΟΣ' column to 'type'
df = df.rename(columns={'ΕΙΔΟΣ': 'type', 'ΘΕΣΗ':  'address', })
```

```python
# Rearrange the columns in the desired order
new_order = ['id', 'name', 'needs_watering', 'is_alive', 'type', 'address', 'latitude', 'longitude', 'zip', 'creation_date', 'last_watered_date']
df = df[new_order]
```

```python
# Get the count of rows
row_count = df.shape[0]

# Display the count of rows
print("Number of rows:", row_count)
```

# Merge with the old trees

```python
# Read the second CSV file into a new DataFrame
df_new = pd.read_csv('tree_old.csv')

# Append the rows of the second DataFrame to the first DataFrame
df_appended = pd.concat([df_new, df], ignore_index=True)
```

```python
# Get the count of rows
row_count = df.shape[0]

# Display the count of rows
print("Number of new trees:", row_count)

# Get the count of rows
row_count = df_new.shape[0]

# Display the count of rows
print("Number of old trees:", row_count)

# Get the count of rows
row_count = df_appended.shape[0]

# Display the count of rows
print("Number of all trees:", row_count)
```

```python
# Save the updated DataFrame to a new CSV file
df_appended.to_csv('tree.csv', index=False)
```
